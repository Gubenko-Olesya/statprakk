## Problem 1

Let $f(x) = 2\sin x - \cos{2x}$. Write a function `beauty_derivative_plot` that takes a point $x$ and plots the function $f(x)$ and its derivative $f'(x)$ in the range $x \in [-2\pi, 2\pi]$. The function should also plot the tangent line to the function at the point $x$.

Example of output for $x = -3, -2, 1.7$:

```python
beauty_derivative_plot(-3)
```

```python
beauty_derivative_plot(-2)
```
<img src="https://i.ibb.co/F4zDq8m0/tangent-line-2-0.png" width="600"/>

```python
beauty_derivative_plot(1.7)
```
<img src="https://i.ibb.co/TBmLXg2h/tangent-line-1-7.png" width="600"/>

````python
import numpy as np
import matplotlib.pyplot as plt

def beauty_derivative_plot(x):

    # Define f(x) and f'(x)
    def f(x):
        return 2 * np.sin(x) - np.cos(2 * x)

    def f_x(x):
        return 2 * np.cos(x) + 2 * np.sin(2 * x)

    # Define the range
    x_vals = np.linspace(-6, 6, 1000)
    y_vals = f(x_vals)
    y_prime_vals = f_x(x_vals)

    # Calculating things
    f_of_x = f(x)
    f_x_of_x = f_x(x)

    tangent = f_x_of_x * (x_vals - x) + f_of_x

    # Plotting
    plt.figure(figsize=(10, 10))
    plt.plot(x_vals, y_vals, label='f(x) = 2sin(x) - cos(2x)', color='green')
    plt.plot(x_vals, y_prime_vals, label="f'(x) = 2cos(x) + 2sin(2x)", color='lime', ls='dashed')
    plt.plot(x_vals, tangent, label=f'tangent at x = {f"{x:.2f}"}', color='saddlebrown')

    plt.axhline(f_of_x, color='peru',linewidth=0.5, ls='dotted')
    plt.axvline(x, color='peru',linewidth=0.5, ls='--')
    plt.scatter(x, f_of_x, color='saddlebrown')
    plt.text(x, f_of_x, f'   f({x:.2f}) = {f_of_x:.2f}', fontsize=10, ha='left', color='saddlebrown')
    plt.title('Function, Derivative and Tangent Line')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.axhline(0, color='black',linewidth=0.5, ls='--')
    plt.axvline(0, color='black',linewidth=0.5, ls='--')
    plt.xlim(-6, 6)
    plt.ylim(-5, 5)
    plt.legend()
    plt.grid()
    plt.show()

beauty_derivative_plot(1.7)
````

- Find which year each species had the largest population. Mark these years in the plot of populations.

````python
import numpy as np
import matplotlib.pyplot as plt

data = np.loadtxt('populations.txt')
years = data[:, 0].astype(int)
hare = data[:, 1]
lynx = data[:, 2]
carrot = data[:, 3]
species = ['Hares', 'Lynxes', 'Carrots']

plt.figure(figsize=(15, 5))
plt.plot(years, hare, label='hares', color='deeppink')
plt.plot(years, lynx, label='lynxes', color='maroon')
plt.plot(years, carrot, label='carrots', color='olive')

plt.plot(years[np.argmax(hare)], max(hare), color = "deeppink", marker = "o", label = f'year of hares largest population = {f"{years[np.argmax(hare)]}"}')
plt.plot(years[np.argmax(lynx)], max(lynx), color = "maroon", marker = "o", label = f'year of lynxes largest population = {f"{years[np.argmax(lynx)]}"}')
plt.plot(years[np.argmax(carrot)], max(carrot), color = "olive", marker = "o", label = f'year of carrots largest population = {f"{years[np.argmax(carrot)]}"}')

plt.title('Population Size of Each Species Over Time')
plt.xlabel('Year')
plt.ylabel('Population Size')
plt.xticks(years)
plt.legend()
plt.grid()
plt.show()
````

- Assuming total population of hares and lynxes is 100%, plot stacked bar graph showing ratio of each species for each year in the period.

````python
total = hare + lynx
hare_ratio = hare / total
lynx_ratio = lynx / total

plt.figure(figsize=(15, 5))
plt.bar(years, hare_ratio, label='hares', color='pink')
plt.bar(years, lynx_ratio, bottom=hare_ratio, label='lynxes', color='maroon')

plt.xticks(years, rotation=90)
plt.xlabel('Year')
plt.ylabel('Population Ratio')
plt.title('Stacked Bar Graph For Hares and Lynxes Populations')
plt.legend()
plt.show()
````

- Find the mean and std of the populations of each species. Plot the historgam of population for each species and show mean values with vertical line. Show each histogram in a separate plot. Arrange 3 plots in a row.

````python
mean_hare = np.mean(hare)
std_hare = np.std(hare)

mean_lynx = np.mean(lynx)
std_lynx = np.std(lynx)

mean_carrot = np.mean(carrot)
std_carrot = np.std(carrot)

plt.figure(figsize = (15,5))

#hares
plt.subplot(1,3,1)
plt.hist(hare, bins = 10, color = 'deeppink', edgecolor='black')
plt.axvline(mean_hare, ls = '--', color = "gold")
plt.text(mean_hare, 4, f' mean: {mean_hare:.2f}', color='gold', fontsize=10, ha='left')
plt.title('Histogram Of Hares Population')
plt.xlabel('Population')
plt.ylabel('Frequency')

#lynxes
plt.subplot(1,3,2)
plt.hist(lynx, bins = 10, color = 'maroon', edgecolor='black')
plt.axvline(mean_lynx, ls = '--', color = "gold")
plt.text(mean_lynx, 4, f' mean: {mean_lynx:.2f}', color='gold', fontsize=10, ha='left')
plt.title('Histogram Of Lynxes Population')
plt.xlabel('Population')
plt.ylabel('Frequency')

#carrots
plt.subplot(1,3,3)
plt.hist(carrot, bins = 10, color = 'olive', edgecolor='black')
plt.axvline(mean_carrot, ls = '--', color = "gold")
plt.text(mean_carrot, 4, f'  mean: {mean_carrot:.2f}', color='gold', fontsize=10, ha='left')
plt.title('Histogram Of Carrots Quantity')
plt.xlabel('Quantity')
plt.ylabel('Frequency')
````

- Find which species (hares or lynxes) has the largest population for each year. Print the result as [H, H, L, H, ...]. Plot a pie chart showing the ratio of "H" and "L" values.

````python
result = np.where(hare > lynx, 'H', 'L')

print(result.tolist(), '\n')

labels = ['Hares', 'Lynxes']
sizes = [np.sum(result == 'H'), np.sum(result == 'L')]
colors = ['pink', 'maroon']

plt.pie(sizes, labels=labels, colors=colors,
        autopct='%1.1f%%', startangle=90)
plt.title('The Ratio of "H" and "L" Values')
plt.show()
````

- Plot the change in the hare population and in the lynx population for each year. Find the correlation coefficient between the both time series.

````python
hare_change = np.diff(hare)
lynx_change = np.diff(lynx)

plt.figure(figsize=(15, 5))

plt.plot(years[1:], hare_change, label='change in hare population', color = "deepskyblue")
plt.plot(years[1:], lynx_change, label='change in lynx population', color = "navy")

plt.xlabel('Years')
plt.ylabel('Change in Population')
plt.title('Change in Hare and Lynx Populations')
plt.legend()
plt.xticks(years)
plt.grid()

print(f"Correlation coefficient between hare and lynx population changes: {np.corrcoef(hare_change, lynx_change)[0,1]}", '\n')
````

- Using a scatter plot, show the population of hares vs lynxes by year (each point corresponds to a particular year, and the point coordinates correspond to the population of the two species in that year).

````python
plt.figure(figsize=(15, 5))
plt.scatter(hare, lynx, color = 'navy')

plt.xlabel('Hare Population')
plt.ylabel('Lynx Population')
plt.title('Hare vs Lynx Population by Years')

for i, year in enumerate(years):
    plt.text(hare[i], lynx[i], f'  {str(year)}', fontsize=10, ha='left', color = 'navy')

plt.grid(True)
plt.tight_layout()
plt.show()
````
